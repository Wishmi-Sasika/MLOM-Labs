```python
filename = '/content/drive/MyDrive/Colab Notebooks/Data/data.txt'
raw_text = open (filename, 'r', encoding = 'utf-8').read()
raw_text = raw_text.lower()
```

```python
# Create mapping of unique chars to integers
chars = sorted (list (set (raw_text)))
print (chars)
char_to_int = dict ((c, i) for i, c in enumerate (chars))
```

```python
print (char_to_int)
```

```python
int_to_char = {i: char for i, char in enumerate (chars)}
```

```python
print (int_to_char)
```

```python
n_chars = len (raw_text)
n_vocab = len (chars)
print ("Total characters : ", n_chars)
print ("Total vocab (unique characters) : ", n_vocab)
```

```python
# Prepare the dataset of input to output pairs encoded as integers
seq_length = 15   # can be changed
                  # including spaces also.
dataX = []
dataY = []

for i in range (0, n_chars - seq_length, 1):
  seq_in = raw_text [i: i + seq_length]
  seq_out = raw_text [i + seq_length]
  dataX.append ([char_to_int[char] for char in seq_in])
  dataY.append ([char_to_int[seq_out]])

n_patterns = len (dataY)
print ("Total patterns : ", n_patterns)
```

```python
print (dataY)
```

```python
print (dataX)
```

```python
import numpy

# Preprocessing

# Reshape X to be [sample, time steps, features]
X = numpy.reshape (dataX, (n_patterns, seq_length, 1))

# Normalize - rescaling the integer values
X = X / float(n_vocab)
print (X)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
```

```python
# One hot encode the output variable
y = to_categorical(dataY)
print(y)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import ModelCheckpoint
```

```python
# Define the LSTM model
model = Sequential()
model.add (LSTM (256, input_shape = (X.shape[1], X.shape[2])))  # it can have 1 or more training samples
model.add (Dropout (0.2))
model.add (Dense (y.shape[1], activation = 'softmax'))
model.compile (loss = 'categorical_crossentropy', optimizer = 'adam')

# Define the checkpoint
filepath = "weights-improvement-{epoch:02d}-{loss:.4f}.hdfs"
checkpoint = ModelCheckpoint (filepath, monitor = 'loss', verbose = 1, save_best_only = True, mode = 'min')
callbacks_list = [checkpoint]
```

```python
# Change the hyperparameter values and train model
epochs = 10
batch_size = 128

model.fit (X, y, epochs = epochs, batch_size = batch_size, callbacks = callbacks_list)
```

```python
#load the network weights
filename="weights-improvement-10-2.4088.hdfs"
model.load_weights(filename)
model.compile(loss='categorical_crossentropy',optimizer='adam')
```
